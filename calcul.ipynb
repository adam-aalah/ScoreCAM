```python
from PIL import Image
import os, sys
import pandas as pd
import cv2
import statistics
import numpy as np
from PIL import Image
data = pd.DataFrame(columns=['name', 'mean', 'median','meand', 'percentile 75','percentile 90','propotion %'])
img_basenames = []
for file in os.listdir("R:/Bureau/punch_gradcam/annot/crop"):
    if file.endswith(".JPEG"):
        img_basenames.append(file)
for basename in img_basenames:
    f_name, f_ext = os.path.splitext(basename)
    name = f_name.replace('_crop','')
    ima = cv2.imread(f'R:/Bureau/punch_gradcam/annot/crop/'+name+'_crop.JPEG')
    im = cv2.imread(f'R:/Bureau/punch_gradcam/result_punching_bag_'+name+'.JPEG')
    d = ima.flatten()
    dd=im.flatten()
    mean = statistics.mean(d)
    mean = mean/255
    meand=statistics.mean(dd)
    meand = meand/255
    median = np.median(d)
    median = median/255
    por = (mean/meand-mean)*100
    seven = np.percentile(ima, 75)
    seven = seven/255
    nine = np.percentile(ima, 90)
    nine = nine/255
    data = data.append(pd.DataFrame(data={'name': name, 'mean': mean, 'median': median,'meand':meand, 'percentile 75': seven,'percentile 90': nine, 'propotion %':por}, index = [0]), ignore_index = True)
        
print('done!')
```
